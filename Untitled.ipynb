```python
import networkx as nx
import numpy as np
from networkx.drawing.nx_pydot import graphviz_layout
```

```python
import numpy as np
from collections import namedtuple


class Node(object):
    def __init__(self, vector):
        self.vector = tuple(vector)
        self.id = (0,0)
    
    def __hash__(self):
        return(hash(self.vector))

    def __eq__(self, other):
        return self.vector == other.vector

class StateTree(object):
    """Holds a tree of potential states, an agent and information about degrees of freedom"""
    
    def __init__(self, degree, depth, vector_length = 3, initial_state = None):
        self.degree = degree
        self.depth = depth
        self.vector_length = vector_length
        self.graph = nx.DiGraph()
        self.node_count = 0
        if initial_state:
            self._root_state = initial_state
        else:
            self._root_state = {
                'time' : 0,
                'action' : None,
                'state' : np.random.rand(self.vector_length)
            }
        # self.current_state = self.root
        self.add_node(None, self._root_state)

    def add_node(self, parent, attributes):
        self.graph.add_node(self.node_count, **attributes)
        if parent:
            self.graph.add_edge(parent, self.node_count)
        self.node_count += 1
    
    def draw_graph(self):
        pos = graphviz_layout(self.graph, "dot")
        labels = {n : self.print_state(n) for n in self.graph}
        nx.draw(self.graph, pos, labels=labels)

    def print_state(self, node):    
        attr = self.graph.nodes[node]
        return "(" + ', '.join([f"{e:.2}" for e in attr['state']]) + ")"
    
    def print_node(self, node):
        attr = self.graph.nodes[node]
        vector_str = "(" + ', '.join([f"{e:.2}" for e in attr['state']]) + ")"
        return f"Time: {attr['time']:}, Action: {attr['action']}, State: {vector_str}"

    def fill_with_random_nodes(self):
        last_row = [0]
        for d in range(1, self.depth):
            new_row = []
            for parent_node in last_row:
                for n in range(self.degree):
                    new_node = self.node_count
                    self.node_count += 1
                    new_row.append(new_node)
                    node_attributes = {
                        'time' : d,
                        'action' : n,
                        'state' : np.random.rand(self.vector_length)
                    }
                    self.graph.add_node(new_node, **node_attributes)
                    self.graph.add_edge(parent_node, new_node)
                    # self.graph.add_node(
                    #     new_node,
                    print("       " * d, self.print_node(new_node))
            last_row = new_row
            

    def node_diff(self, node0, node1):
        state0 = self.graph.nodes[node0]['state']
        state1 = self.graph.nodes[node1]['state']
        return abs(np.linalg.norm(node1 - node0))
    
```

```python
tree = StateTree(3, 3)
```

```python
tree.fill_with_random_nodes()
```

```python
tree.draw_graph()
```
